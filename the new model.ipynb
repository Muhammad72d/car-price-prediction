```python
#بسم الله 
import numpy as np 
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
data=pd.read_csv('car_price_prediction.csv')
```

```python
data['Levy']=data['Levy'].str.replace('-','0').astype(int)
print(data['Levy'][23])
```

```python
data=data.dropna()
print(data.columns)
```

```python
#cleaning data
data['Mileage']=data['Mileage'].str.replace('km','').astype(float)
data['Engine volume']=data['Engine volume'].str.replace('Turbo','').astype(float)
################
#petrol = 0 ,hybrid = 1 ,diesel = 2 ,  CNG  = 3 ,  LPG  = 4 ,plug in 1 = 5 ,hydrogen = 6
data['Fuel type']=data['Fuel type'].str.replace('Petrol','0').str.replace('Hybrid','1').str.replace('Diesel','2').str.replace('CNG','3').str.replace('LPG','4').str.replace('Plug-in 1','5').str.replace('Hydrogen','6').astype(float)
################
data['Leather interior']=data['Leather interior'].str.replace('Yes','1').str.replace('No','0').astype(int)
data['Gear box type']=data['Gear box type'].str.replace('Manual','0').str.replace('Automatic','1').str.replace('Tiptronic','2').str.replace('Variator','3').astype(int)
data['Wheel']=data['Wheel'].str.replace('Left wheel','0').str.replace('Right-hand drive','1').astype(int)
data['Doors']=data['Doors'].str.replace('-May','').str.replace('-Mar','').str.replace('>','').astype(int)
data['Drive wheels']=data['Drive wheels'].str.replace('Front','20').str.replace('Rear','02').str.replace('4x4','22').astype(int)
data['Category']=data['Category'].str.replace('Coupe','0').str.replace('Goods wagon','1').str.replace('Hatchback','2').str.replace('Jeep','3').str.replace('Limousine','4').str.replace('Microbus','5').str.replace('Minivan','6').str.replace('Pickup','7').str.replace('Sedan','8').str.replace('Universal','9').str.replace('Cabriolet','10').astype(int)
```

```python
le=LabelEncoder()
data['Model']=le.fit_transform(data['Model'])
data['Manufacturer']=le.fit_transform(data['Manufacturer'])
data['Color']=le.fit_transform(data['Color'])
```

```python
y=data['Price']
data.drop(columns='Price')
```

```python
#splites a section four testing
x_train,x_test,y_train,y_test=train_test_split(data,y,test_size=0.15,random_state=40,shuffle=True)

print(y_test.shape)
```

```python
y_test= y_test.values.reshape(-1,1)
```

```python
model=LinearRegression(fit_intercept=True)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred[:5])
print(y_test[:5])
```

```python
mae=mean_absolute_error(y_test,y_pred,multioutput='uniform_average')
```

```python
y_pred=pd.DataFrame(y_pred)
```

```python
print(y_pred.shape)
print(y_test.shape)
```

```python
mae=mean_squared_error(y_test,y_pred)
print(mae)
```
